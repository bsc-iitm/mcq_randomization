```python
import pandas as pd
from datetime import datetime, timedelta
import time
import io
import json
import numpy as np
import random
```

```python
## Get the last modified time
dtime = datetime. now() + timedelta(seconds=3)
unixtime = time. mktime(dtime. timetuple())
```

```python
## Create default
def_dict = {'all_or_nothing_grading': False,
 'defaultFeedback': '',
 'last_modified': unixtime,
 'multiple_correct': False,
 'multiple_selections': False,
 'show_answer_when_incorrect': False,
 'tags': [],
 'type': 0,
 'version': '1.5'}
```

```python

##Read Question spreadsheet
questions_6_7 = pd.read_csv("sample.csv")

#Create Choice arrays
choice_6 = questions_6_7.iloc[:,1:6]
choice_7 = questions_6_7.iloc[:,6:11]
```

```python
## All fucntions

def convert(o):
    if isinstance(o, np.int64): return int(o)  
    raise TypeError

def question_generator(**kwargs):
  """
  -- Arguments :
  qnum - question number
  text - text entity to be added to the question
  week_start - imdb dataset week start
  week_end - imdb dataset week end
  url - imdb dataset url
  """
  qnum = kwargs.get('qnum', None)
  text = kwargs.get('text', None)

  #q1 = f''
  #q2 = f''
  #q3 = f''
  #q4 = f''
  #q5 = f''
  q6 = f'On gadm.org, download the shapefile for "{text}" and look at the level 1 DBF file. In the shapefile what type of data does the "NAME_1" column hold?'
  q7 = f'On gadm.org, download the shapefile for "{text}" and look at the level 1 DBF file. In the shapefile what type of data does the "HASC_1" column hold?'
  #q8 = f''
  #q9 = f''
  #q10 = f''
  #q_value = [q1,q2,q3,q4,q5,q6,q7,q8,q9]
  #q_key =[1,2,3,4,5,6,7,8,9] 
  q_value = [q6,q7]             # Feed the number of questions to be generated. Fill appropriate questions in the section above
  q_key =[6,7]                  # Feed the number of questions to be generated. Fill appropriate questions in the section above
  q_dict = dict(zip(q_key,q_value))
  question = q_dict[qnum]
  return question

def create_mcq_json(qnum,question_input_array,choice_array,qdesc,outputfilename,qcount):

  
  final_json = {"question_type": "0"}
  q_list = []
  for i in range(0,qcount):
    row_json = def_dict
    if qnum == 9 :
      row_dict = question_input_array.loc[i]
      city = row_dict[0]
      row_json["question"] = question_generator(qnum = qnum,text = city )
    else:
      row_dict = question_input_array.loc[i]
      item = row_dict[0]
      row_json["question"] = question_generator(qnum = qnum,text = item )
    row_json['description'] = qdesc + "-" + str(i)
    #row_json["question"] = q7.format(city)
    choice_list = []
    n = choice_array.shape[1]
    for j in range(0,n):
      choices = choice_array.loc[i]
      choice_dict = {'feedback':''}
      if j == 0:
        if (qnum == 7 or qnum == 1):
          choice_dict["score"] = float(1)  ## To modify the score for a particular question add the qnum to if condition and modify the score to required value
        else:
          choice_dict["score"] = float(1)
      else: 
        choice_dict["score"] = float(0)
      choice_dict["text"] = str(choices[j])
      choice_list.append(choice_dict)
    random.shuffle(choice_list)
    row_json["choices"] = choice_list
    q_list.append(row_json.copy()) 
  final_json["questions"]= q_list

 

  with io.open(outputfilename, 'w', encoding='utf-8') as f:
    f.write(json.dumps(final_json, indent = 4,default=convert,ensure_ascii=False))

  return final_json
```

```python
#create mcq jsons
create_mcq_json(6,questions_6_7,choice_6,'TDS-W5-Q6','TDS_W1_Q6.json',10)
create_mcq_json(7,questions_6_7,choice_7,'TDS-W5-Q7','TDS_W1_Q7.json',10)

```
